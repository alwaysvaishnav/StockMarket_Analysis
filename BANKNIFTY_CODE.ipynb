```python
import csv
import pandas as pd
import numpy as np
from collections import OrderedDict
```

```python
banknifty=pd.read_csv("BANKNIFTY.csv")
```

```python
banknifty.head()
```

```python
banknifty_unique_date=banknifty['DATE'].unique()
```

```python
len(banknifty_unique_date)
```

```python
group_high = banknifty.groupby('DATE')

df1 = group_high.apply(lambda x: x['HIGH'].unique())
```

```python
group_low = banknifty.groupby('DATE')

df2 = group_low.apply(lambda y: y['LOW '].unique())
```

```python
df1.head()
```

```python
df2.head()
```

```python
print(len(df1.keys()))
print(len(df2.keys()))
```

```python
final=pd.DataFrame(df1.items(), columns=['Date', 'HIGH'])
final2=pd.DataFrame(df2.items(), columns=['Date', 'LOW'])
```

```python
final.head(10)
```

```python
for ind in final.index:
    final['HIGH'][ind]=max(final['HIGH'][ind])
```

```python
final2.head(10)
```

```python
ar=[]
for ind in final2.index:
    ar.append(min(final2['LOW'][ind]))
```

```python
final['LOW']=ar
```

```python
final.head(20)
```

```python
final['Date'] = pd.to_datetime(final['Date'], format='%d-%m-%Y')
```

```python
final=final.sort_values(by=['Date'], ascending=True)
```

```python
final.head(20)
```

```python
final.to_csv('DAILY_RANGE_BANKNIFTY.csv', header=True, index=False)
```
